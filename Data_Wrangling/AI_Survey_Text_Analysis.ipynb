```python
from datetime import date
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob, Word
from wordcloud import WordCloud

file = "/home/shared/AI_Attitudes.csv"
survey = pd.read_csv(file)
survey
```

## Data Processing

The code below renames columns to one or two works and drops columns that aren't relevant to our analysis.

```python
# Type your code here

survey.columns=['ID', 'Start', 'End', 'Email', 'Name', 'Last Edit', 'Age', 'Gender', 'Excited', 'Why Excited', 'Concerned', 'Why Concerned', 'Use Freq', 'How Used', 'Tools', 'Report']
survey.drop(columns=['Email', 'Name', 'Last Edit', 'Report'], inplace=True)
survey
```

## Create a TextBlob

The code below creates a TextBlob for answers to question "What makes you excited about the increasing use of AI?".

A TextBlob is like a string, but with more features.

```python
# Put all the responses in the 'Why Excited' column into a TextBlob called why_excited.
all_responses = ""
response_list = survey['Why Excited'].dropna()
for item in response_list:
    if type(item)==type("abc"):
        all_responses = all_responses + "\n" + item
why_excited = TextBlob(all_responses)
why_excited.correct()
why_excited
```

## Perform Some Analysis

The following code performs some analysis using `textblob` features. Working with textblobs rather that strings can be useful for a number of reasons:
- You can correct spelling mistakes to make responses more consistent
- You can get sentiment scores that give a numerical representation of, for example, how positive or negative the language is
- You can search for word frequencies without worrying about differences due to capitalization
- You can search for word frequencies based not words and their common synonyms.

```python
# Sentiment Polarity. +1 = very positive sentiments, -1 = very negative sentiments
why_excited.sentiment.polarity
```

```python
# Sentiment Subjectivity. +1 = Highly subjective, 0 = Highly objective
why_excited.sentiment.subjectivity
```

```python
# Count how many times a word occurs - in this case the word 'research'
word = Word("research")
why_excited.words.count(word)
```

```python
# Create a list of synonyms for the word 'research'. We can then search not just for 'research' but for its synonyms as well.
synonyms = set()
for item in word.synsets:
    for w in item.lemma_names():
        synonyms.add(w)
synonyms
```

```python
# Now we will iterate through our list of synonyms so that we can count how many times a word or its synonym appear
count = 0
for w in synonyms:
    count = count + why_excited.words.count(w)
print(count)
```

## Create a Word Cloud

Using the Python `wordcloud` library, you can create a simple visualization to illustrate the frequencies of different words.

```python
# Create a word cloud using responses to "Why Excited" question
# The word cloud libary tasks strings, so we need to convert our TextBlob back into a string
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(str(why_excited))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
