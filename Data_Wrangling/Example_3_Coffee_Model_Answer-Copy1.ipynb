### Load Raw HTML Table into Dataframe

The code below will load the table into a dataframe called `coffee`.

```python
# Import libraries
import requests as r, pandas as pd, matplotlib.pyplot as plt, numpy as np

# Create a header that says the request is coming form a browser-like agent (this is to prevent the website blocking our request)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
}

# Make an http request to get the webpage with the following url
url = "https://worldpopulationreview.com/country-rankings/coffee-producing-countries"
page = r.get(url, headers = headers)
tables = pd.read_html(page.content)
coffee = tables[0]
coffee
```

### Rename Columns

There should be four columns. Rename them as:

- 'Not used'
- 'Country'
- '2022 Production (tonnes)'
- '2022 Yield (kg/hectare)'

```python
# Type your code here

# Rename the columns
coffee.columns = ['Not used', 'Country', '2022 Production (tonnes)', '2022 Yield (kg/hectare)']
coffee
```

### Drop Column and Row

Drop the first column (called 'Not Used') and the last row (the 'Total' row).

```python
#Type your code here

# Drop column 1
coffee = coffee.drop(columns=['Not used'])
# Drop last row (row 78)
coffee = coffee.drop(index=78)
coffee
```

### Create Columns

Create a 'Prod (kt)' column and a '% of Global' column. These columns will both be floats and default values can be 0.0 for now. 

```python
# Type your code here

# Create a new column called 'Units' with the default value '-' (string)
coffee['% of Global']=0.0
# Create a new column called 'Prod (kt)' with the default value 0.0 (float)
coffee['Prod (kt)']=0.0
coffee
```

### Calculate Numerical Value of Production

In this step (which is the hardest!) you need to tak the string value of the production and convert it into a numerical value to put in the 'Prod (kt)' column. The 'Prod (kt)' column should contain the correct numerical value for each country's production in kt, based on the information available. You will need to take into account that the 'M' and 'K' values are strings but tell you the units.

```python
# Type your code here
num_rows = coffee.shape[0]
for i in range(num_rows):
    if("M" in coffee.iloc[i, 1]):
        coffee.iloc[i, 4] = float(coffee.iloc[i, 1].replace("M" , ""))*1000
    elif("K" in coffee.iloc[i, 1]):
        coffee.iloc[i, 4] = float(coffee.iloc[i, 1].replace("K", ""))
    else:
        coffee.iloc[i, 4] = float(coffee.iloc[i, 1])/1000
coffee
```

### Calculate Percentage of Global

We now want to work out what each country's production is as a percentage of the total production and put this in the '% of Global' column.

```python
# Type your code here

total_prod = coffee['Prod (kt)'].sum()
coffee['% of Global'] = coffee['Prod (kt)'] / total_prod*100
coffee
```
