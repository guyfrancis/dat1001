```python
import numpy as np
from PIL import Image

# 1. Open the image file
img = Image.open('star_icon.png') # Replace 'your_image.jpg' with your image file path

# 2. Convert the image object to a NumPy array
img_array = np.array(img) # np.asarray(img) also works

# Optional: Print the array's shape and data type
print(f"Image shape: {img_array.shape}")
print(f"Data type: {img_array.dtype}")
```

```python
for x in range(32):
    for y in range(32):
        for z in range(0, 3):
            if(img_array[x][y][z]==255):
                img_array[x][y][z]=0
            elif(img_array[x][y][z]==0):
                img_array[x][y][z]=255
```

```python
from matplotlib import pyplot as plt
plt.imshow(img_array, interpolation='nearest')
plt.show()
```
