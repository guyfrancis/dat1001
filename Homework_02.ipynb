Firstly, grab the CSV file from the following URL path if you have not done so already. The file is called 'USD_GBP_EUR_SF_5YR.csv'.


```python
import numpy as np
file = "USD_GBP_EUR_SF_5YR.csv"
np.set_printoptions(floatmode='fixed', precision=3)
c1 = np.loadtxt(file, skiprows=1, usecols = (1, 2, 3), delimiter=",")
print(c1.shape)
c2 = c1.transpose()
print(c2.shape)
c3 = np.reshape(c2, (3, 60))
print(c3.shape)
```

```python
print(c3)
```

```python
eur_to_usd = c3[0]
```

```python
print(eur_to_usd)
```
