- Row 1: Monthly exchange rates between Euros (EUR) and US Dollars (USD) for the past 5 years. Each value represents how many USD to 1 EUR. The first value in the array is the oldest (Jan 2021).
- Row 2: Monthly exchange rates between British Pounds (GBP) and US Dollars (USD) for the past 5 years. Each value represents how many USD to 1 GBP. The first value in the array is the oldest (Jan 2021).
- Row 3: Monthly exchange rates between US Dollars (USD) and Swiss Francs (SF) for the past 5 years. Each value represents how many SF to 1 USD. The first value in the array is the oldest (Jan 2021).

````python
import numpy as np
file = "USD_GBP_EUR_SF_5YR.csv"
np.set_printoptions(floatmode='fixed', precision=3)

# Load data into a numpy array called c2
c1 = np.loadtxt(file, skiprows=1, usecols = (1, 2, 3), delimiter=",")
print(c1.shape)

# c2 is c1 with rows and columns flipped
c2 = c1.transpose()
print(c2.shape)

# c3 is c2 reshaped into 3 rows of 60 months (12 x 5) of data
c3 = np.reshape(c2, (3, 60))
print(c3.shape)

# Print the whole array
print(c3)
````
