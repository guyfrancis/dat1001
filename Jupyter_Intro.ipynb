## Formatting Markdown Cells

You can format the content of markdown cells.

For example, If you start a line with a # in a markdown cell, it becomes an heading.

# This is a heading

Two ## symbols give a second-level heading and three ### symbols give a third-level heading.

## Second-level heading

### Third-level heading

In most enviroments, double clicking on a markdown cell allows you to edit the content. To see the rendered content, you can press **SHIFT+ENTER**.

## More Markdown Formats

We have seen you can have different headings in the markdown text.

You can also surround text with two stars to make it bold:

**This text is bold**

You can surround text with one star to make it italics

*This is italics*

If you want a bulleted list, you can use the - symbol:

- item 1
- item 2
- item 3

If you want a numbered list, write the number followed by a period and a space:

1. Item 1
2. Item 2
3. Item 3

If you want to illustrate some code in the markdown text, you need to start and end with three back-quotes (```):

```
print("Hello World!')
name = input("What's your name?")
print("Hello" + name)
```

If you want to illustrate a code snippet in line use single back-quotes: `print("Hello World!")`.

You can put hyperlinks in markdown text. Use square brackets for the link title and put the url in regular parentheses immediately after:
