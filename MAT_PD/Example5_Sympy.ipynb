## Using sympy and Numerical Methods Example

In this notebook, we use the Python library `sympy` to symbolically differentiate a function. We then use a numerical method involving calculus to find a root.

Run the code cells below and check the output. The first code cell sets everything up.

```python
from sympy import symbols
import matplotlib.pyplot as plt
import numpy as np

x = symbols('x')
def newton(f, a0, tol=1e-7, max_iter=100):

    a = a0
    step = 0
    for _ in range(max_iter):
        print(f"Step {step}, x-value: {a:.10f}")
        fa = f.subs(x, a)
        dfa = f.diff(x).subs(x, a)
        iter_pairs.append((a, fa))
        if abs(dfa) < 1e-10: # Avoid division by zero
            break
            
        a_new = a - fa / dfa
        
        if abs(a_new - a) < tol: # Convergence check
            return a_new
        
        a = a_new
        step+=1
    return a
print("Setup complete.")
```

```python
# Now define a function and display it nicely
f = x**2 - 1
f
```

```python
# Now differentiate the function
df = f.diff('x')
df
```

```python
# Now use newton function to find a root
start_value = 5
iter_pairs = []
root = newton(f, start_value)
print(f"Root: {root:.10f}")
```

```python
# Finally, let's viualize what we have done with a plot
left_lim = -5
right_lim = 5
low_lim = -5
high_lim = 20

x_vals = np.arange(left_lim, right_lim, 0.2)
y_vals = []
for num in x_vals:
    y_vals.append(f.subs(x, num))

fig, ax = plt.subplots()
ax.plot(x_vals, y_vals)
ax.plot([left_lim, right_lim], [0,0], color='black')
ax.plot([0,0], [low_lim, high_lim], color='black')
for i in range(len(iter_pairs)-1):
    ax.plot([iter_pairs[i][0], iter_pairs[i][0]], [0, iter_pairs[i][1]], 'r--o')
    ax.plot([iter_pairs[i][0], iter_pairs[i+1][0]],[iter_pairs[i][1], 0], 'r--o')
plt.title("Illustration of Iteration")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
