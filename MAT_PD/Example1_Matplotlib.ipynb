## Example 2

Click on the cell below and press **SHIFT+ENTER** to run the code.

Then try modifying the code to see how the output changes.

```python
x = np.arange(0, 2*np.pi,  0.2)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y, 'o-b')
plt.plot(x, z, 'D-.y')
plt.show()
```

## Example 3

Click on the cell below and press **SHIFT+ENTER** to run the code.

Then try modifying the code to see how the output changes.

These data are taken from [https://en.wikipedia.org/wiki/List_of_countries_with_highest_military_expenditures](https://en.wikipedia.org/wiki/List_of_countries_with_highest_military_expenditures).

See if you can add in more countries to the visualizations.

```python
# Data
countries = ['US', 'China', 'Russia', 'Germany', 'India']
spending = [997, 314, 149, 88.5, 86.1]

# Create bar graph
plt.title("Defence Spending - Top 5 Countries (2024)")
plt.xlabel("Spending (billions USD)")
plt.ylabel("Country")
plt.barh(countries, spending, color='red')
plt.show()

# Create pie chart
plt.title("Defence Spending - Top 5 Countries (2024)")
plt.pie(spending, labels=countries, autopct='%1.1f%%')
plt.show()
```
