# Using Meteostat

In this activity, we use a custom library to import weather data for 3 Colorado locations from the web and then plot some scatter plots and box plots.

The weather data come from the [Meteostat](www.meteostat.net) website. Follow the link to learn more about Meteostat. If you want to look at data for othe weather stations around the world, you look up their station codes on this website.

```python
# Import required libraries
from datetime import date
import meteostat as ms, numpy as np, pandas as pd, matplotlib.pyplot as plt

# Set display options for numpy
np.set_printoptions(floatmode='fixed', precision=1)

# Set start and end date for data
start = date(2025, 1, 1)
end = date(2025, 12, 31)

# Select Station(s) - the numbers in the parentheses are the station codes
denver_stn = ms.stations.meta('72469')
alamosa_stn = ms.stations.meta('72462')
lamar_stn = ms.stations.meta('KLAA0')

# Get daily data and store in pandas dataframe for each location
denver = ms.daily(denver_stn, start, end).fetch()
alamosa = ms.daily(alamosa_stn, start, end).fetch()
lamar = ms.daily(lamar_stn, start, end).fetch()
```

## Part 1: Box Plots
Let's compare the 3 locations using a side-by-side box plot.

```python
# Create a combined plot for the daily temperature range for all three locations
plt.xlabel("1 = Alamosa, 2 = Denver, 3 = Lamar")
plt.ylabel("Wind Speed (km/h)")
plt.title("Box Plot of Average Daily Windspeeds for 3 CO Locations (2025)")
plt.boxplot([alamosa['wspd'], denver['wspd'], lamar['wspd']], showmeans=True)
plt.show()
```

## Part 2: Scatter Plot

Let's do a scatter plot of windspeed vs air pressure for one of the locations.

```python
# Create a scatter plot of windspeed vs air pressure for Alamosa
import matplotlib.pyplot as plt
x1 = alamosa['pres']
y1 = alamosa['wspd']
plt.xlabel('Air Pressure (mb)')
plt.ylabel('Wind Speed (kmh)')
plt.title('Daily Avg Air Pressure vs Wind Speed in Alamosa, 2025')
plt.scatter(x1, y1)
plt.show()
```

## Part 3: Correlation

It looks like there is a reasonably strong negative correlation between air pressure and windspeed, so let's work out the correlation coefficient and line of best fit.

```python
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(x1, y1, nan_policy='omit')
print(f"slope: {slope:.3f}")
print(f"intercept: {intercept:.3f}")
print(f"r: {r:.3f}")

x = np.arange(x1.min(), x1.max(), 10)
plt.xlabel('Air Pressure (mb)')
plt.ylabel('Wind Speed (kmh)')
plt.title('Daily Avg Air Pressure vs Wind Speed in Alamosa, 2025')
plt.scatter(x1, y1)
lbl=f"w = {slope:.3f}p + {intercept:.3f}"
plt.plot(x, slope*x+intercept, 'r:', label=lbl)
plt.legend()
plt.show()
```

## Part 4: Temperature over Time

Finally, let's plot graphs of the daily temperatures over time for one of the locations.

```python
plt.plot(denver['temp'], label='Average')
plt.plot(denver['tmax'], label='Max')
plt.plot(denver['tmin'], label='Min')
plt.xlabel("Date")
plt.ylabel("Temp (deg C)")
plt.title("Daily Temps in Denver (2025)")
plt.legend()
plt.show()
```
