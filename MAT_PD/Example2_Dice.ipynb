# Sum of Dice Experiment

In this workbook, you will simulate an experiment where you roll a fixed number of dice a certain number of times. In class we did an experiment where we rolled 2 dice repeatedly and recorded the sum of the scores on the two dice on each roll. We saw that the most likely total score was 7 and scores of 2 and 12 were the least common. 

This notebook generalizes that experiment, allowing you to simulate rolling `n` dice `t` times.

## Simulating More Dice

In the following code cells, we simulate rolling `n` dice `t` times and working out the total score as before.

For example, if you roll 3 dice twice, the total on each set of rolls could be any whole number between 3 and 18. You might get results that look like this:

- First roll: 2, 3, 1 giving a total score of 6
- Second roll: 6, 2, 5 giving a total score of 13

## Your Task
Run the code below and view the histogram that is generated from the data. The variables are initially set to rolling 2 dice 100 times.

Try different values for `n` and `t` to see what you get. What patterns do you see? Try a few different numbers of dice (n = 2, 3, 4, 5, ...) and for each number of dice, try t = 100, 1000, 10000, 100000.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# set values for n and t
n = 2
t = 100

# Generate psuedo-random data
score_low = n
score_high = 6*n
rng = np.random.default_rng()
total_scores = []
for roll in range(t):
    dice_scores = rng.integers(low=1, high=7, size=n)
    total_scores.append(int(dice_scores.sum()))

# Peint the first few values in total_scores
print(total_scores[0:10])
```

```python
# Create a customized histogram
plt.hist(total_scores, bins = np.arange(score_low - 0.5, score_high + 1.5, 1), alpha=0.8, color='green')
plt.title(f'Score Distribution Rolling {n} Dice {t} Times')
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

## Comparison to Normal Distribution

The following code generates the same histogram, but also plots the normal distribution curve whose mean and standard deviation are the same as that of the simulated data you have generated. How good an approximation is the normal distribution for different values of `n` and `t`?

```python
# This code repeats the histogram but also plots the theoretical
# normal distribution that would model the data

ts = np.array(total_scores)
m = ts.mean()
s = ts.std(ddof=1)
x = np.arange(n-2, 6*n+2, 0.5)
y = np.exp(-0.5*((x-m)/s)**2)/(s*math.sqrt(2*math.pi))*t
plt.plot(x, y, color='r')
plt.hist(total_scores, bins = np.arange(score_low - 0.5, score_high + 1.5, 1), alpha=0.8, color='green')
plt.title(f'Score Distribution Rolling {n} Dice {t} Times')
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
## Calculating Probabilities
import scipy.stats as stats

lower_score = 10
upper_score = 25

# probability score is between lower_score and upper_score based on our simulation
in_range_count = len(np.where((lower_score < ts) & (ts < upper_score))[0])
in_range_count
total_count = t
print("Probability based on experimental data:")
print(f"Probability score is between {lower_score} and {upper_score} exclusive is {in_range_count/total_count:.3f}")
print()

# Now estimate probability with normal distribution (using a continuity correct)
prob = stats.norm.cdf(x=upper_score-0.5, loc=m, scale = s) - stats.norm.cdf(x=lower_score+0.5, loc=m, scale = s)
print("Probability based on normal distribution:")
print(f"Probability score is between {lower_score} and {upper_score} exclusive is {prob:.3f}")
print()
```
