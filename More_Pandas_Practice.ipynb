```python
data = {"Country": ["Brazil", "Indonesia", "Sao Tome and Principe"], "Production":['3.2M', '794.8K', '8']}
```

```python
import pandas as pd
import numpy as np
cd = pd.DataFrame(data)
cd
```

```python
cd['Units']="-"
cd['Prod (kt)']=0.0
```

```python
cd
```

```python
num_rows = cd.shape[0]
for i in range(num_rows):
    if("M" in cd.iloc[i, 1]):
        cd.iloc[i, 2] = 'M'
        cd.iloc[i, 3] = float(cd.iloc[i, 1].replace("M" , ""))
    elif("K" in cd.iloc[i, 1]):
        cd.iloc[i, 2] ="K"
        cd.iloc[i, 3] = float(cd.iloc[i, 1].replace("K", ""))
    else:
        cd.iloc[i, 3] = float(cd.iloc[i, 1])
```

```python
cd
```

```python
# Now update numerical value based on units
for i in range(len(cd['Production'])):
    if(cd.iloc[i, 2]=="M"):
        cd.iloc[i, 3] *=1000
    elif(cd.iloc[i, 2]=="-"):
        cd.iloc[i, 3] /= 1000
```

```python
cd
```

```python
cd. describe()
```

```python
# Now we can answer the key questions
print(f"Total global coffee production is {cd['Prod (kt)'].sum():.2f} kt")
print(f"Mean coffee production for coffee-producing countries is {cd['Prod (kt)'].mean():.2f} kt")
```

```python
print(f"Biggest production for one country {cd['Prod (kt)'].max():.2f} kt")
```
