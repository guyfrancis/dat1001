## What are complex numbers?

In mathematics, complex numbers are an extension of the more familiar real number system. They allow for the possibility of square roots of negative numbers. In particular, the square root of -1 is denoted j. 

$$j = \sqrt{-1}$$

A complex number $z$ has two parts, the real part and the imaginary part. We can write a complex number as 

$$z = a + bj$$

where $a$ is the real part and $b$ is the imaginary part. Here $a$ and $b$ are real numbers.

Python includes complex numbers as a fundamental datatype (unlike most programming languages). 

Let's create some examples of complex numbers in Python. We'll use variable names z, u, v, w for complex numbers.

First we need to import the cmath library that will make using complex numbers easier.

```python
# The cmath library allows 
import cmath
```

```python
# Create a complex number with real part = 3 and imaginary part = 2
z = 3+2j
```

```python
# Create a complex number with real part = 4 and imaginary part = -5
w = 4-5j
```

```python
# We can add, subtract, multiply and divide complex numbers to get new complex numbers
print(z+w, z-w, 2*z, z*w, w/z)
```

## Mathematical operations
Look at the output above. Can you see how each expression is obtained? Adding, subtracting and multiplying by a scalar (just a real number) is fairly straightforward. However, it is less obvious what is going on with multiplication and division.

$$ z+w= (a+p) + (b+q)j$$
