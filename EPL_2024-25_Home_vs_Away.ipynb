```python
# Import 2024-25 EPL results from a CSV file into a NumPy array, reshape the data and convert to a 1 dimensional list

import numpy as np
file = "EPL Results 2024-25.csv"
raw_table = np.loadtxt(file, usecols=range(1, 21), skiprows=1, delimiter=",", dtype=str)
raw_table = np.reshape(raw_table, (1, 400))
raw_list = (raw_table.tolist())[0]
raw_list = [item for item in raw_list if item != '—']
# The raw list is a list of scores from games with the home team score FIRST and away team score SECOND.
# The data type of each element in the raw_list is string
print(raw_list)
```

```python
# We will now convert each string into a tuple with first value equal to home goals and second to away goals.
# We will put the tuples into a list called 'scores'
scores = []
for item in raw_list:
    str_vals = item.split('–')
    this_tuple = int(str_vals[0]), int(str_vals[1])
    scores.append(this_tuple)
print(scores)
```

```python
# We will now do some analysis on the list of scores
# We will count how many games resulted in a home win, tie and away win
hwin = 0
tie = 0
awin = 0
for x in scores:
    if x[0]>x[1]:
        hwin+=1
    elif x[0]==x[1]:
        tie+=1
    else:
        awin+=1

# Finally, we will print the number of home wins, ties and away wins and the percentage of each
games = len(scores)
print("Games:", games)
print("Home wins:", hwin)
print("Ties:", tie)
print("Away win", awin)
print(f"Home win percentage:{hwin/games*100:.1f}%")
print(f"Tie percentage:{tie/games*100:.1f}%")
print(f"Away win percentage:{awin/games*100:.1f}%")
```
