````python
# Type your code here

def wind_categories(wspd_list):
    from collections import Counter
    speed_cats = ["High wind days" if spd > 15 else "Normal wind days" if spd >=5 else "Low wind days" for spd in wspd_list]
    count = Counter(speed_cats)
    return dict(count)

wind_categories(avg_wind_speeds)
````
