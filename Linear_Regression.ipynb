## Creating and Plotting Bivariate Data

Now let's add some data and plot it. The data below gives the goals scored (x) and goals conceded (y) for 20 teams in the English Premier League during the 2024-25 season.

Read the code carefully and then run it. The result should be a scatter plot of goals scored versus goals conceded.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([86, 69, 72, 64, 68, 58, 58, 66, 58, 66, 54, 51, 42, 46, 44, 54, 64, 33, 36, 26])
y=np.array([41, 34, 44, 43, 47, 51, 46, 59, 46, 57, 54, 51, 44, 62, 54, 69, 65, 80, 82, 86])

plt.scatter(x, y)
plt.show()
```

## Computing the Correlation Coefficient (r) and Regression Line

Now we'll work out the equation of the least squares regression line. To do this, we will use the `stats` package. We will also calculate the correlation coefficient between goals conceded and goals scored. 

Run the code below to calculate the slope and intercept for our regression model. Once these have been calculated, the code prints out the slope, intercept and correlation coefficient to 3 decimal places.

What does the r-value (correlation coefficient) tell you?

```python
from scipy import stats

m, b, r, p, std_err = stats.linregress(x, y)
print(f"slope (m): {m:.3f}")
print(f"intercept (b): {b:.3f}")
print(f"Correlation coefficient (r): {r:.3f}")
print(f"Standard error: {std_err:.3f}")
```

## Line of Best Fit

Let's plot the scatter plot and the regression line on the same axes to see what that looks like. Run the code below to do this. Here is what the different variables represent

- `x` is the independent variable, goals scored
- `y` is the depedendent variable, goals conceded
- `y_hat` are the predicted y-values based on the x-values

The mathematical model to predict goals conceded is:
$$
\hat{y} = m\cdot x + b
$$

```python
y_hat = m*x+b
plt.plot(x, y_hat)
plt.scatter(x, y)
plt.xlabel("Goals scored")
plt.ylabel("Goals conceded")
plt.title("Goals scored vs goals conceded by team, EPL 2024-25")
plt.show()
```

## Predicting Values

## Residuals

The residuals are the differences between actual y-values and predicted y-values. In mathematical notation, the $i^{th}$ residual $e_i$ is given by the formula

$$
e_{i} = y_i - \hat{y}_i
$$

The residuals tell us a few things about the model:

- **Model Accuracy**: They show the magnitude of prediction errors, helping determine if the model is a good fit.
- **Linearity Check**: If residuals in a plot show a curve rather than a random scatter, it indicates the relationship is non-linear and a linear model is inappropriate.
- **Constant Variance (Homoscedasticity)**: Residuals should be randomly scattered. If they fan out (a cone shape), it indicates heteroscedasticity, meaning the model's accuracy varies across the range of data.
- **Identify Outliers**: Unusually large positive or negative residuals indicate data points that the model cannot explain well.
- **Independence**: In time series, patterns in residuals can indicate that errors are correlated over time (autocorrelation).

Below we create the residuals and plot them against the original x values.

```python
# Create residuals
residuals = y-y_hat

# Plot residuals
plt.scatter(x, residuals)
plt.plot([x.min(), x.max()], [0,0])
for i in range(len(residuals)):
    plt.plot([x[i], x[i]], [0, residuals[i]], ':b')
plt.title("Residual Plot for Model (Goals conceded vs goals scored)")
plt.xlabel("Goals scored")
plt.ylabel("Residual: y - y_hat")
plt.show()
```

## Extension: Calculating m, b and r from Scratch

For reference, we include the calculations required to work out the model parameters m and b (slope and intercept) and the correlation coefficient r.

Check that the code below gives you the same values as above for m, b and r.

```python
# Calculate m b and r from scratch

import math

n = len(x)
Sx = x.sum()
Sy = y.sum()
Sxy = (x*y).sum()
Sxx = (x*x).sum()
Syy = (y*y).sum()

r = (n*Sxy - Sx*Sy)/(math.sqrt(n*Sxx - (Sx)**2)*math.sqrt(n*Syy - (Sy)**2))
print(f"Correlation coefficient (r) = {r:.3f}")

m = (n*Sxy - Sx*Sy)/(n*Sxx - (Sx)**2)
print(f"Slope (m) = {m:.3f}")

b = (Sy-m*Sx)/n
print(f"Intercept (b) = {b:.3f}")
```
