## Creating and Plotting Bivariate Data

Now let's add some data and plot it. The data below gives the goals scored (x) and goals conceded (y) for 20 teams in the English Premier League during the 2024-25 season.

### Task 2
Read the code carefully and then run it. The result should be a scatter plot of goals scored versus goals conceded.

```python
import matplotlib.pyplot as plt

x=[86, 69, 72, 64, 68, 58, 58, 66, 58, 66, 54, 51, 42, 46, 44, 54, 64, 33, 36, 26]
y=[41, 34, 44, 43, 47, 51, 46, 59, 46, 57, 54, 51, 44, 62, 54, 69, 65, 80, 82, 86]

plt.scatter(x, y)
plt.show()
```

## Labeling Plots

### Task 3

Let's make the scatter plot easier to interpret. The following lines should add x and y labels and a title to the plot. The title has been left blank. Add the title "Goals score vs goals conceded by team, EPL 2024-25" so that this displays as well as the axis labels.

```python
plt.xlabel("Goals scored")
plt.ylabel("Goals conceded")
plt.title("")
plt.scatter(x, y)
plt.show()
```

## Computing the Regression Line

Now we'll work out the equation of the least squares regression line. To do this, we need to use the stats package. We will also calculate the correlation coefficient between goals conceded and goals scored. 

### Task 4

Run the code below to calculate the slope and intercept for our regression model. Once these have been calculated, the code prints out the slope, intercept and correlation coefficient to 3 decimal places. See if you can modify the code so that it prints out these values to 2 decimal places.

What does the r-value (correlation coefficient) tell you?

```python
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x, y)
print(f"slope: {slope:.3f}")
print(f"intercept: {intercept:.3f}")
print(f"r: {r:.3f}")
```

### Task 5

Let's plot the scatter plot and the regression line on the same axes to see what that looks like. Run the code below to do this.

```python
def predict(x):
  return slope * x + intercept
    
model_data = list(map(predict, x))
plt.plot(x, model_data)
plt.scatter(x, y)
plt.xlabel("Goals scored")
plt.ylabel("Goals conceded")
plt.title("Goals scored vs goals conceded by team, EPL 2024-25")
plt.show()
```

## Using the Predict Function

We can use the model to make predictions about number of goals conceded as a function of goals scored. For example we might like to know how many goals a team that has scored 80 goals might concede during a season.

The predict function created above uses the linear model to do this.
