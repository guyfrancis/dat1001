See if you can answer the questions below. 
1. What was the highest temperature during the 5-year period in deg F and when did it occur?
2. What was the lowest temperature during the 5-year period in deg F and when did it occur?
3. What was the average temperature over the 5-year period?
4. What day in the 5-year period had the smallest temperature range, and what was it, in deg F?
5. (Challenge) Find the maximum temperature range in each year 2021, 2022, 2023, 2024 and 2025. Do these all occur roughly around the same time of year?

```python
# Load Denver temperature data and set print options
import numpy as np
file = "Denver_Daily_Temps_2021-25.csv"
den_temps = np.loadtxt(file, usecols=(1, 2, 3), skiprows=1, delimiter=",")
np.set_printoptions(floatmode='fixed', precision=1)
den_temps = den_temps.T
print(den_temps)
```

```python
# Convert to deg F and split into 3 separate arrays
den_temps = 9/5*den_temps + 32
tavg = den_temps[0]
tmin = den_temps[1]
tmax = den_temps[2]
```

```python
# Maximum temp over 5 years
max_temp = tmax.max()
max_temp_days = np.where(tmax == max_temp)
print(f"Max temp was {max_temp:.2f} on day {max_temp_days[0][0]}.")
```

```python
# Minimum temp over 5 years
min_temp = tmin.min()
min_temp_days = np.where(tmin == min_temp)
print(f"Min temp was {min_temp:.2f} on day {min_temp_days[0][0]}.")
```

```python
# Average temp over 5 years
ave_temp = tavg.mean()
print(f"Average temp over 5 years was {ave_temp:.2f}.")
```

```python
# Finding max daily temperature range
trange = tmax - tmin
print(f"Maximum daily temp range was {trange.max():.2f}")
when_max_range = np.where(trange == trange.max())
print(f"This occurred on day {when_max_range[0][0]}, which was day {when_max_range[0][0] - 365*3 - 366} of year 2025.")
```

```python
# Challenge: finding max temperature range in each year
# As the output indicates, this tends to occur early in the year during winter (but not always)
yr = 2021
yr_start = 0
while yr <= 2025:
    print("Year:", yr)
    yr_end = yr_start+365
    if yr==2024:
        yr_end+=1
    print(f"Max temperature range in {yr} was {trange[yr_start:yr_end].max():.2f} def F.")
    when = np.where(trange[yr_start:yr_end]==trange[yr_start:yr_end].max())[0][0]
    print("This occurred on day:", when)
    yr+=1
    yr_start = yr_end
```
