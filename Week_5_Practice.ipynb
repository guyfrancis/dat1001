## Task 1b

Now open the file in this Jupyter notebook and load it into a pandas dataframe.

You will need to import pandas first. Then you will need to use the `pd.read_csv()` function to read it into a dataframe. Call your dataframe `usi` for US inflation. Then uses the `pd.describe()` method to answer the following questions:

1. What was the mean annual inflation rate between 1960 and 2024?
2. What was the highest annual inflation rate?
3. What was the lowest annual inflation rate?

You might also find it helpful to rename the column 'FPCPITOTLZGUSA' to 'Annual Inflation %'.

```python
# Type your code here

import pandas as pd

inflation_file = 'Inflation.csv'

usi = pd.read_csv(inflation_file)
usi.rename(columns={'FPCPITOTLZGUSA':'Annual Inflation %'}, inplace=True)
usi.describe()
```

# Answers to Questions
1. Mean inflation rate was 3.75%
2. Highest annual rate was 13.55%
3. Lowest annual rate was -0.36% (deflation)

## Task 1c

Now import `maplotlib` and create a boxplot of the annual inflation rate for the period 1960-2024.

```python
import matplotlib.pyplot as plt

plt.boxplot(usi['Annual Inflation %'])
plt.xticks([1], ['UK'])
plt.ylabel('Percent')
plt.title("US Inflation 1960-2024")
plt.show()
```

## Task 2

Now repeat the above tasks for the annual inflation rate for the UK for a similar time period (1956-2023) by downloading the data from this URL:

http://fred.stlouisfed.org/series/CPALTT01GBA657N

Upload the data to Jupyter Hub and load into a pandas dataframe in this notebook.

Use these data and the US data above to plot a side-by-side box plot showing the annual inflation rates in the US and UK on the same graph, for the period 1960-2023.

What do the boxplots tell you about US vs UK inflation over this period?

```python
# Type your code here

inflation_file = 'UK_inflation.csv'
uki = pd.read_csv(inflation_file)
uki.rename(columns={'CPALTT01GBA657N':"Annual Inflation %"}, inplace=True)
uki.describe()
```

```python
plt.boxplot([usi['Annual Inflation %'][0:65], uki['Annual Inflation %'][4:]])
plt.xticks([1, 2], ['US', 'UK'])
plt.ylabel('Percent')
plt.title("US vs UK Inflation")
plt.show()
```

The box plots show that UK inflation was higher on average and the UK had more years of very high inflation.
