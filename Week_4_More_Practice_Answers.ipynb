## Exercise 3

Create a **horizontal bar chart** for the data given below. The data give the solar electricity generation in TeraWatt-hours for the top ten countries, plus the total for the rest of the world.

Label your chart as follows:

- Title = "Solar Power Generation By Country 2024"
- y-axis = "Country"
- x-axis = "Generation (TWh)"

```python
countries = ['Rest of World', 'South Korea', 'Italy', 'Australia', 'Spain', 'Germany', 'Brazil', 'Japan', 'India', 'United States', 'China']
values = [413.13, 32.73, 35.81, 49.84, 58.6, 70.99, 74.68, 102.02, 133.81, 303.17, 834.1]

# type your code here

plt.barh(countries, values)
plt.title("Solar Power Generation By Country (2024)")
plt.xlabel("Generation (TWh)")
plt.ylabel("Country")
plt.show()
```

## Exercise 4

Create a **pie chart** for the data given below. The data give the solar electricity generation in TeraWatt-hours for the top five countries, 
plus the total for the rest of the world.

Give the pie chart the title "Solar Power Generation By Country 2024". Include the percentages in the chart, appropriately formatted.

```python
countries = ['China', 'United States', 'India', 'Japan', 'Brazil', 'Rest of World']
values = [834.1, 303.17, 133.81, 102.02, 74.68, 661.1]

# Type your code here
plt.pie(values, labels=countries, autopct='%1.1f%%')
plt.title("Solar Power Generation By Country 2024")
plt.show()
```

## Exercise 5

The data below given the average daily temperatures for London and Denver for January 2026.

Use these data to create side-by-side box plots to visualize the distribution of temperatures.

Your box plot should be appropriately labeled and titled:

- x-labels: "London" and "Denver"
- y-label: "Temperature (deg C)"
- Title: "Average Daily Temperatures January 2026"

Use the box plots to describe two differences between the average daily temperatures in London and Denver during this time.

```python
london_temps = [3.2, 2.0, 0.7, -0.4, -0.9, -0.2, 3.8, 4.2, 4.2, 1.5, 4.0, 
                10.4, 9.0, 4.8, 8.4, 6.7, 8.0, 8.4, 8.9, 8.7, 7.8, 8.5, 
                7.4, 7.3, 6.9, 5.4, 7.3, 5.4, 4.1, 7.4, 7.8]
denver_temps = [6.0, 6.8, 6.7, 9.2, 5.8, 4.9, 4.7, 1.5, -1.9, 
                -0.9, 2.2, 4.4, 6.6, 6.5, 6.1, -1.0, -3.7, 1.9, 
                -0.2, 1.7, 0.6, -2.7, -8.8, -10.7, -10.1, -5.9, 
                -0.0, 1.3, 1.7, 1.0, 3.2]

plt.boxplot([london_temps, denver_temps])
plt.title("Average Daily Temperature January 2026")
plt.xticks([1, 2], ['London', 'Denver'])
plt.ylabel("Temp (deg C)")
plt.show()
```

Difference 1: The average temperature was higher in London

Difference 2: There is much more variation in temperature in Denver
