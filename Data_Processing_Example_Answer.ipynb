```python
ulim = 500
numlist = []
import random as r
for x in range(ulim):
    numlist.append(r.randint(40, 100))
print(numlist)

for i in range(len(numlist)):
    if numlist[i]<70:
        numlist[i]+=15
```

```python
def lettergrade(num):
    if num >= 90:
        return "A"
    elif num >= 80:
        return "B"
    elif num >= 70:
        return "C"
    elif num >=60:
        return "D"
    else:
        return "F"
```

```python
grades = []
for x in numlist:
    grades.append(lettergrade(x))
print(grades)
```

```python
a_grades = 0
b_grades = 0
c_grades = 0
d_grades = 0
f_grades = 0

for x in grades:
    if x == "A":
        a_grades+=1
    elif x == "B":
        b_grades+=1
    elif x == "C":
        c_grades+=1
    elif x == "D":
        d_grades+=1
    else:
        f_grades+=1

grade_dist = {}
grade_dist["A"]=a_grades
grade_dist["B"]=b_grades
grade_dist["C"]=c_grades
grade_dist["D"]=d_grades
grade_dist["F"]=f_grades

print(grade_dist)
```
