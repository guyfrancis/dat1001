## Exercise 3

1. Create a list called `LS` containing the elements `'p'`, `3.14`, and `False`. Print out the entire list.

## Exercise 4
1. Create a set A containing the elements 0 and 1.
2. Add the elements 2 and 3 to the set A and print the whole set.
3. Create a set B containing the elements 3, 4, 5 and 6.
4. Remove the element 6 from the set B and print the whole set.
5. Print the set of elements that are in A and B.
6. Print the set of elements that are in B but not in A.

````python
# Answers to Exercise 4
A = {0, 1}
print("1.", A)
A.add(2)
A.add(3)
print("2.", A)
B = {3, 4, 5, 6}
print("3.", B)
B.remove(6)
print("4.", B)
print(B)
print("5.", B.intersection(A)) # print(A.intersect(B)) also works.
print("6.", B.difference(A))
````

## Extension Problems

### Part 1

We noted that tuples and strings are immutable. This means you cannot change them. For example, the follow code won't work.
```
my_tuple = (4, -3, 2)
my_tuple[1] = 7
```
On the other hand, lists and sets are mutable. You can modify them. So the following code does work.
```
my_list = [2, 4, 7, 9. -10]
my_list[2] = 5
```
**Your task**: Do some research to learn about the pros and cons of data structures being mutable or immutable.

### Brief Answer to Part 1

It is perhaps more obvious why a data structure is mutable: Mutable data structures can be easily modified in code. The strongest argument for immutability is security as it places restrictions on what can be changed by code. For example, a password string should not be able to be modified easily in code. There are other arguments for immutability relating to performance issues and efficiency of memory management, but from a programmers point of view, it doesn't really make a big different.

### Part 2

It is common in data science to have nested dictionaries. Consider the example below.
```
d1 = { 'person 1': {'Name':'Alice', 'Age': 37, 'Interests': ['Data Science', 'Coding']},
          'person 2': {'Name':'Bob', 'Age': 42, 'Interests': ['Data Science', 'Cooking']}}
```
This is a dictionary that contains two key-value pairs. The first key is `person 1` and the second key is `person 2`. The value corresponding to each key is another dictionary, giving information about the person.

**Your task**: Run the code to store the dictionary in the variable `d1`. You should be able to copy and paste to do this. Use indexing to print each of the following.
1. The whole dictionary corresponding to `person 1`.
2. The name of `person 2`.
3. The whole list of interests of `person 1`.
4. The second interest of `person 2`.

````python
# Answer Extension Part 2
d1 = { 'person 1': {'Name':'Alice', 'Age': 37, 'Interests': ['Data Science', 'Coding']},
          'person 2': {'Name':'Bob', 'Age': 42, 'Interests': ['Data Science', 'Cooking']}}
print("1.", d1['person 1'])
print("2.", d1['person 2']['Name'])
print("3.", d1['person 1']['Interests'])
print("4.", d1['person 2']['Interests'][1])
````

### Part 3

Suppose the variable mystr stores the string `'Karma'`. How could you change the second character to an `'o'` and store the new string in the original variable? Your solution should work in general with any string: given a string, a new character and an index value, replace the character at the index position with the new character and store in the same variable.

````python
# Answer to Extension Part 3
mystr = 'Karma'
print(mystr)
mystr = mystr[0:1]+'o'+mystr[2:]
print(mystr)
````
