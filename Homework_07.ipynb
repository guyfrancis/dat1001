```python
# Import required libraries
import requests as r, pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np, re

# Create a header that says the request is coming form a browser-like agent (this is to prevent the website blocking our request)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
}
# Make an http request to get the webpage with the following url
url = "https://sports.yahoo.com/nba/standings/"
page = r.get(url, headers = headers)
print(page.status_code)
tables = pd.read_html(page.content)
```

```python
# Store table[0] in a pandas dataframe called 'east' (for Eastern Conference) and do some data cleaning
east = tables[0].copy()
east.drop([6, 11], inplace=True)
east.head()
for x in range(east.shape[0]):
    raw_name = east.iloc[x, 0]
    if x < 9:
        east.iloc[x, 0] = raw_name[1:len(raw_name)]
    elif x == 9:
        east.iloc[x, 0] = raw_name[2:len(raw_name)]
east.rename(columns={'Eastern':'Team'}, inplace=True)
east.drop(columns=['GB', 'Home', 'Div', 'Conf', 'Last 10', 'Odds'], inplace=True)
east['Games'] = east['W']+east['L']
east
```

# Task 2: Points For vs Points Against

Let's look at the relationship between points for (PF) and points against (PA). Note that in this NBA table, points for and against are given as averages per game rather than as totals over all the games.

Run the code below that calculates the correlation coefficient, regression parameters and plots a scatter plot with a line of best fit and a residual plot. Then comment on the output in the markdown cell below.

```python
# Pick the relevant columns of the dataframe and convert to numpy arrays
x = np.array(east['PF'])
y = np.array(east['PA'])

# Calculate regression parameters/statistics
m, b, r, p, std_err = stats.linregress(x, y)
print(f"Slope: {m:.3f}")
print(f"Intercept: {b:.3f}")
print(f"r: {r:.3f}")

# Plot scatter plot and line of best fit
y_hat = m*x+b
plt.plot(x, y_hat)
plt.title("Scatter Plot of Points For vs Points Against (Eastern NBA Conf)")
plt.xlabel("Points For (Per Game)")
plt.ylabel("Points Against (Per Game)")
plt.scatter(x, y)
plt.show()

# Create residuals
residuals = y-y_hat

# Plot residuals
plt.scatter(x, residuals)
plt.plot([x.min(), x.max()], [0,0])
for i in range(len(residuals)):
    plt.plot([x[i], x[i]], [0, residuals[i]], ':b')
plt.title("Residual Plot For Model (Points For vs Points Against)")
plt.xlabel("Points For (Per Game)")
plt.ylabel("Residual: y-y_hat")
plt.show()
```
